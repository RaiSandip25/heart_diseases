EDA for heart diseases with histogram

Load the heart diseases data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
heart_disease_data = pd.read_csv("heart.csv")
```

Heatmap to find the correlation

```python
plt.figure(figsize=(10,8))
correlation = heart_disease_data.corr()
sns.heatmap(correlation, cmap='coolwarm', annot=True)
plt.title("Heatmap of correlation")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=heart_disease_data, x='age', bins=15, kde=True, hue='target')
plt.title("Histogram of Age by Target")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data=heart_disease_data, x='chol', bins=15, kde=True, hue='target')
plt.title("Histogram of Cholesterol by Target")
plt.xlabel("Cholesterol")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.countplot(data=heart_disease_data, x='sex', hue='target', palette='Spectral')
plt.title("Bar Chart of Sex vs Target")
plt.xlabel("Sex (0: Female, 1: Male)")
plt.ylabel("Count")
plt.legend(title="Target (0: No Heart Disease, 1: Heart Disease)")
plt.show()
```
